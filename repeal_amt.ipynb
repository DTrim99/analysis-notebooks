```python
from policyengine_us import Microsimulation
from policyengine_core.reforms import Reform
import numpy as np
import pandas as pd
import os
```

```python
class repeal_salt_reform(Reform):
    def apply(self):
        self.neutralize_variable("alternative_minimum_tax")
```

```python
baseline = Microsimulation(
    dataset="hf://policyengine/policyengine-us-data/enhanced_cps_2024.h5"
)
```

```python
reformed_income = Microsimulation(
    dataset="hf://policyengine/policyengine-us-data/enhanced_cps_2024.h5",
    reform=repeal_salt_reform,
)
```

```python
year_results = []
for year in range(2026, 2035):
    baseline_results = baseline.calculate("household_net_income", period=year)
    reformed_results = reformed_income.calculate("household_net_income", period=year)
    difference = reformed_results - baseline_results
    result = {"year": year, "result": difference.sum() / 1e9}
    year_results.append(result)
    print(f"Year {year} result: {result['result']}")

year_results = pd.concat(year_results)

year_results.to_csv("alternative_minimum_tax_results.csv", index=False)
```
