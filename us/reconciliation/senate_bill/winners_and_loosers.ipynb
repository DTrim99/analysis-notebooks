```python
df = pd.DataFrame({
    'baseline_gross_income': baseline_gross_income,
    'baseline_market_income': baseline_market_income,
    'baseline_net_income': baseline_income,
    'reformed_net_income': reformed_income,
    'unweighted_net_income_change': reformed_income - baseline_income,
    'weighted_net_income_change': (reformed_income - baseline_income) * weight,
    'weighted_population': household_size * weight
})
```

```python
df['pct_change'] = np.where(
    df['baseline_net_income'] != 0,
    ((df['reformed_net_income'] - df['baseline_net_income']) / np.abs(df['baseline_net_income'])) * 100,
    0
)
```

```python
brackets = {
    'Overall': df.index,
    '$1M+ gross income': df[df['baseline_market_income'] >= 1_000_000].index,
    '<$75k gross income': df[df['baseline_market_income'] < 75_000].index,
    '<$25k gross income': df[df['baseline_market_income'] < 25_000].index
}
```

```python
def get_all_breakdowns(indices):
    subset = df.loc[indices]
    
    # Get totals for each weighting
    total_weighted = subset['weights'].sum()
    total_unweighted = len(subset)
    total_population = subset['weighted_population'].sum()
    
    if total_unweighted == 0:
        return {
            'weighted': {'Gaining >5%': 0, 'Gaining <5%': 0, 'Not affected': 0, 'Losing <5%': 0, 'Losing >5%': 0},
            'unweighted': {'Gaining >5%': 0, 'Gaining <5%': 0, 'Not affected': 0, 'Losing <5%': 0, 'Losing >5%': 0},
            'population_weighted': {'Gaining >5%': 0, 'Gaining <5%': 0, 'Not affected': 0, 'Losing <5%': 0, 'Losing >5%': 0}
        }
    
    # Define the groups
    gaining_big = subset['pct_change'] > 5
    gaining_small = (subset['pct_change'] > 0) & (subset['pct_change'] <= 5)
    not_affected = subset['pct_change'] == 0
    losing_small = (subset['pct_change'] < 0) & (subset['pct_change'] >= -5)
    losing_big = subset['pct_change'] < -5
    
    return {
        'weighted': {
            'Gaining >5%': (subset[gaining_big]['weights'].sum() / total_weighted) * 100 if total_weighted > 0 else 0,
            'Gaining <5%': (subset[gaining_small]['weights'].sum() / total_weighted) * 100 if total_weighted > 0 else 0,
            'Not affected': (subset[not_affected]['weights'].sum() / total_weighted) * 100 if total_weighted > 0 else 0,
            'Losing <5%': (subset[losing_small]['weights'].sum() / total_weighted) * 100 if total_weighted > 0 else 0,
            'Losing >5%': (subset[losing_big]['weights'].sum() / total_weighted) * 100 if total_weighted > 0 else 0
        },
        'unweighted': {
            'Gaining >5%': (len(subset[gaining_big]) / total_unweighted) * 100,
            'Gaining <5%': (len(subset[gaining_small]) / total_unweighted) * 100,
            'Not affected': (len(subset[not_affected]) / total_unweighted) * 100,
            'Losing <5%': (len(subset[losing_small]) / total_unweighted) * 100,
            'Losing >5%': (len(subset[losing_big]) / total_unweighted) * 100
        },
        'population_weighted': {
            'Gaining >5%': (subset[gaining_big]['weighted_population'].sum() / total_population) * 100 if total_population > 0 else 0,
            'Gaining <5%': (subset[gaining_small]['weighted_population'].sum() / total_population) * 100 if total_population > 0 else 0,
            'Not affected': (subset[not_affected]['weighted_population'].sum() / total_population) * 100 if total_population > 0 else 0,
            'Losing <5%': (subset[losing_small]['weighted_population'].sum() / total_population) * 100 if total_population > 0 else 0,
            'Losing >5%': (subset[losing_big]['weighted_population'].sum() / total_population) * 100 if total_population > 0 else 0
        }
    }
```

```python
df['weights'] = weight

# Calculate breakdowns for each bracket
results = {}
for bracket_name, indices in brackets.items():
    results[bracket_name] = get_all_breakdowns(indices)

```

```python
print("Tax Reform Impact Breakdown")
print("=" * 80)

for bracket in brackets.keys():
    print(f"\n{bracket}:")
    
    print("\n  Weighted (by household):")
    for category, percentage in results[bracket]['weighted'].items():
        print(f"    {category}: {percentage:.1f}%")
    
    print("\n  Unweighted:")
    for category, percentage in results[bracket]['unweighted'].items():
        print(f"    {category}: {percentage:.1f}%")
    
    print("\n  Population weighted:")
    for category, percentage in results[bracket]['population_weighted'].items():
        print(f"    {category}: {percentage:.1f}%")
```
