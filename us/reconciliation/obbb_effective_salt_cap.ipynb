```python
def create_situation_with_two_axes(
    household_type="married",
    include_child=False,
    mortgage_interest=0,
    charitable_donations=0,
):
    """Create situation dictionary with two axes (reported_salt and employment_income)

    Parameters:
    -----------
    household_type : str
        Either "single", "married", or "head_of_household"
    include_child : bool
        Whether to include a child in the household
    mortgage_interest : float
        Amount of mortgage interest to include
    charitable_donations : float
        Amount of charitable cash donations to include
    """

    situation = {
        "people": {},
        "families": {},
        "marital_units": {},
        "tax_units": {},
        "spm_units": {},
        "households": {},
        "axes": [
            [
                {
                    "name": "reported_salt",
                    "count": 700,
                    "min": -50000,
                    "max": 250000,
                    "period": 2026,
                }
            ],
            [
                {
                    "name": "employment_income",
                    "count": 1400,
                    "min": 0,
                    "max": 1000000,
                    "period": 2026,
                }
            ],
        ],
    }

    # Add people based on household type
    situation["people"]["you"] = {"age": {"2026": 40}}

    if household_type == "married":
        situation["people"]["spouse"] = {"age": {"2026": 40}}

    if include_child:
        situation["people"]["child"] = {"age": {"2026": 10}}

    # Build members list
    members = ["you"]
    if household_type == "married":
        members.append("spouse")
    if include_child:
        members.append("child")

    # Set up family and other units
    situation["families"]["your family"] = {"members": members}

    if household_type == "married":
        situation["marital_units"]["your marital unit"] = {"members": ["you", "spouse"]}
    else:
        situation["marital_units"]["your marital unit"] = {"members": ["you"]}

    # Set up tax unit with appropriate filing status
    situation["tax_units"]["your tax unit"] = {
        "members": members,
        "state_and_local_sales_or_income_tax": {"2026": 0},
    }

    # Add optional person-level parameters for mortgage interest and charitable donations to the head
    if mortgage_interest > 0:
        situation["people"]["you"]["mortgage_interest"] = {"2026": mortgage_interest}

    if charitable_donations > 0:
        situation["people"]["you"]["charitable_cash_donations"] = {
            "2026": charitable_donations
        }

    situation["spm_units"]["your household"] = {"members": members}

    situation["households"]["your household"] = {
        "members": members,
        "state_code": {"2026": "CA"},
    }

    return situation
```

```python
def calculate_marginal_rate(group):
    """Calculate the marginal property tax rate for a group"""
    # Check if group needs sorting
    if not group["reported_salt"].is_monotonic_increasing:
        group = group.sort_values("reported_salt")

    # Calculate differences in income tax and property tax
    group["income_tax_diff"] = group["income_tax"].diff()
    group["reported_salt_diff"] = group["reported_salt"].diff()

    # Calculate the marginal rate (change in income tax / change in property tax)
    group["marginal_property_tax_rate"] = (
        -group["income_tax_diff"] / group["reported_salt_diff"]
    )

    # Handle edge cases
    group["marginal_property_tax_rate"] = group["marginal_property_tax_rate"].replace(
        [np.inf, -np.inf, np.nan], 0
    )

    return group
```

```python
def process_effective_cap_data(df):
    """Process the data by calculating marginal rates for each income level"""
    # Sort the dataframe
    sorted_df = df.sort_values(by=["policy", "employment_income", "reported_salt"])

    # Calculate the marginal property tax rate for each income level
    processed_df = (
        sorted_df.groupby(["policy", "employment_income"])
        .apply(calculate_marginal_rate)
        .reset_index(drop=True)
    )

    return processed_df
```

```python
def calculate_effective_salt_cap_over_earnings(
    household_type="married",
    include_child=False,
    mortgage_interest=0,
    charitable_donations=0,
):
    """Calculate effective SALT cap for varying property tax and income levels"""
    # Create situation with specified parameters
    situation = create_situation_with_two_axes(
        household_type=household_type,
        include_child=include_child,
        mortgage_interest=mortgage_interest,
        charitable_donations=charitable_donations,
    )

    # Create dataframe to store results
    result_df = pd.DataFrame()

    # Run simulation for Full OBBB reform
    simulation_obbb = Simulation(situation=situation, reform=obbb_full_reform)
    employment_income = simulation_obbb.calculate(
        "employment_income", map_to="household", period=2026
    )
    reported_salt = simulation_obbb.calculate(
        "reported_salt", map_to="household", period=2026
    )
    income_tax = simulation_obbb.calculate(
        "income_tax", map_to="household", period=2026
    )
    salt_deduction = simulation_obbb.calculate(
        "salt_deduction", map_to="household", period=2026
    )

    # Create DataFrame with Full OBBB data
    obbb_full_df = pd.DataFrame(
        {
            "employment_income": employment_income,
            "reported_salt": reported_salt,
            "income_tax": income_tax,
            "salt_deduction": salt_deduction,
            "policy": "Full OBBB reform",
        }
    )

    # Run simulation for OBBB no salt reform
    simulation_obbb_no_salt = Simulation(
        situation=situation, reform=obbb_no_salt_reform
    )
    employment_income = simulation_obbb_no_salt.calculate(
        "employment_income", map_to="household", period=2026
    )
    reported_salt = simulation_obbb_no_salt.calculate(
        "reported_salt", map_to="household", period=2026
    )
    income_tax = simulation_obbb_no_salt.calculate(
        "income_tax", map_to="household", period=2026
    )
    salt_deduction = simulation_obbb_no_salt.calculate(
        "salt_deduction", map_to="household", period=2026
    )

    # Create DataFrame with OBBB no salt data
    obbb_with_salt_uncapped_df = pd.DataFrame(
        {
            "employment_income": employment_income,
            "reported_salt": reported_salt,
            "income_tax": income_tax,
            "salt_deduction": salt_deduction,
            "policy": "OBBB with SALT uncapped",
        }
    )

    # Run simulation for baseline reform
    simulation_baseline = Simulation(
        situation=situation, reform=baseline_brachning_reform
    )
    employment_income = simulation_baseline.calculate(
        "employment_income", map_to="household", period=2026
    )
    reported_salt = simulation_baseline.calculate(
        "reported_salt", map_to="household", period=2026
    )
    income_tax = simulation_baseline.calculate(
        "income_tax", map_to="household", period=2026
    )
    salt_deduction = simulation_baseline.calculate(
        "salt_deduction", map_to="household", period=2026
    )

    # Create DataFrame with baseline data
    baseline_df = pd.DataFrame(
        {
            "employment_income": employment_income,
            "reported_salt": reported_salt,
            "income_tax": income_tax,
            "salt_deduction": salt_deduction,
            "policy": "Current Law",
        }
    )

    # Combine all dataframes
    result_df = pd.concat([obbb_full_df, obbb_with_salt_uncapped_df, baseline_df])

    return result_df
```

```python
def create_max_salt_line_graph(df, threshold=0.1, title_suffix=""):
    """Create a line graph showing the maximum SALT values where marginal_property_tax_rate > threshold"""
    # Define colors
    BLUE = "#3378b2"
    LIGHT_GRAY = "#bababa"
    DARK_GRAY = "rgba(74, 74, 74, 0.9)"

    # Filter data where marginal_property_tax_rate > threshold
    filtered_df = df[df["marginal_property_tax_rate"] > threshold]

    # For each unique policy and employment income, find the maximum SALT value
    max_salt_by_income = (
        filtered_df.groupby(["policy", "employment_income"])["reported_salt"]
        .max()
        .reset_index()
    )

    # Create figure
    fig = go.Figure()

    # Define colors for different policies
    colors = {
        "OBBB with SALT uncapped": BLUE,
        "Current Law": LIGHT_GRAY,
        "Full OBBB reform": DARK_GRAY,
    }

    # Add lines for each policy
    for policy in max_salt_by_income["policy"].unique():
        policy_data = max_salt_by_income[
            max_salt_by_income["policy"] == policy
        ].sort_values("employment_income")

        # Include all policies
        fig.add_trace(
            go.Scatter(
                x=policy_data["employment_income"],
                y=policy_data["reported_salt"],
                mode="lines",
                line=dict(color=colors.get(policy, "#000000"), width=1.5),
                name=policy,
                hovertemplate="Income: $%{x:,.0f}<br>Effective SALT Cap: $%{y:,.0f}<extra></extra>",
            )
        )

    # Update layout
    fig.update_layout(
        title=f"Effective SALT Cap by Income Level{title_suffix}",
        xaxis_title="Employment Income",
        yaxis_title="Effective SALT Cap (2026)",
        xaxis=dict(
            tickformat="$,.0f",
            showgrid=True,
            gridcolor="rgba(0,0,0,0.1)",
            range=[0, 1000000],
        ),
        yaxis=dict(
            tickformat="$,.0f",
            showgrid=True,
            gridcolor="rgba(0,0,0,0.1)",
            range=[0, 160000],
        ),
        plot_bgcolor="white",
        legend=dict(
            x=1, y=0.99, bgcolor="rgba(255,255,255,0.8)", bordercolor="rgba(0,0,0,0.1)"
        ),
    )

    # Apply additional formatting if format_fig function is defined
    try:
        fig = format_fig(fig)
    except:
        pass

    return fig
```

```python
def generate_salt_cap_graph(
    household_type="married",
    include_child=False,
    mortgage_interest=0,
    charitable_donations=0,
    title_suffix="",
):
    """Generate a single SALT cap graph with the specified parameters and display it immediately"""

    print(f"Calculating data for: {title_suffix}...")

    # Calculate data
    result_df = calculate_effective_salt_cap_over_earnings(
        household_type=household_type,
        include_child=include_child,
        mortgage_interest=mortgage_interest,
        charitable_donations=charitable_donations,
    )

    # Process data
    processed_df = process_effective_cap_data(result_df)

    # Create and display the figure
    fig = create_max_salt_line_graph(processed_df, title_suffix=title_suffix)
    fig.show()

    return processed_df, fig
```
