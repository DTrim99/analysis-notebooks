```python
years = range(2026, 2035)
```

```python
all_average_changes = []
all_relative_changes = []
```

```python
# Calculate impacts for each year
for year in years:
    print(f"Calculating for year {year}...")
    
    baseline = Microsimulation(dataset="hf://policyengine/policyengine-us-data/enhanced_cps_2024.h5")
    reformed = Microsimulation(reform=reform, dataset="hf://policyengine/policyengine-us-data/enhanced_cps_2024.h5")
    
    # baseline_net_income = baseline.calculate("household_net_income", map_to="household", period=year)
    baseline_income_tax = baseline.calculate("income_tax", map_to="household", period=year)
    baseline_benefits = baseline.calculate("household_benefits", map_to="household", period=year)
    baseline_medicaid = baseline.calculate("medicaid", map_to="household", period=year)
    baseline_aca = baseline.calculate("aca_ptc", map_to="household", period=year)
    baseline_chip = baseline.calculate("per_capita_chip", map_to="household", period=year)
    baseline_income = (baseline_income_tax * -1) + baseline_medicaid + baseline_aca + baseline_chip + baseline_benefits
    
    baseline_net_income = baseline.calculate("household_net_income_including_health_benefits", map_to="household", period=year)

    count_people = baseline.calculate("household_count_people", period=year)
    household_weight = baseline.calculate("household_weight", period=year)
    weighted_income = MicroSeries(
        baseline_net_income, weights=household_weight * count_people
    )
    decile = weighted_income.decile_rank().values
    
    household_income_decile = (np.where(baseline_net_income < 0, -1, decile)).astype(int)
    
    reformed_income_tax = reformed.calculate("income_tax", map_to="household", period=year)
    reformed_benefits = reformed.calculate("household_benefits", map_to="household", period=year)
    reformed_medicaid = reformed.calculate("medicaid", map_to="household", period=year)
    reformed_aca = reformed.calculate("aca_ptc", map_to="household", period=year)
    reformed_chip = reformed.calculate("per_capita_chip", map_to="household", period=year)
    reformed_income = (reformed_income_tax * -1) + reformed_medicaid + reformed_aca + reformed_chip + reformed_benefits
    
    income_change = reformed_income - baseline_income
    
    # Calculate metrics for this year
    average_change = income_change.groupby(household_income_decile).mean()
    # For relative change: use absolute value approach similar to tax calculation
    baseline_sum = baseline_income.groupby(household_income_decile).sum()
    reformed_sum = reformed_income.groupby(household_income_decile).sum()
    change_sum = income_change.groupby(household_income_decile).sum()
    
    # Initialize relative change with zeros
    relative_change = pd.Series(np.zeros(len(baseline_sum)), index=baseline_sum.index)
    
    # Calculate relative change only where baseline is non-zero
    mask = baseline_sum != 0
    relative_change[mask] = (change_sum[mask] / np.abs(baseline_sum[mask])) * 100
    
    relative_change = relative_change.round(1)
    
    # Store results (excluding negative income decile)
    all_average_changes.append(average_change[average_change.index > 0])
    all_relative_changes.append(relative_change[relative_change.index > 0])
```

```python
# Create DataFrames for each year
yearly_results = []
for i, year in enumerate(years):
    df_year = pd.DataFrame({
        f'Average change {year}': all_average_changes[i].values,
        f'Relative change {year}': all_relative_changes[i].values,
        'Income decile': all_average_changes[i].index
    })
    yearly_results.append(df_year.set_index('Income decile'))
```

```python
# Combine all years into one DataFrame
combined_df = pd.concat(yearly_results, axis=1)

# Calculate mean across all years
mean_average_change = combined_df[[col for col in combined_df.columns if 'Average change' in col]].mean(axis=1)
mean_relative_change = combined_df[[col for col in combined_df.columns if 'Relative change' in col]].mean(axis=1)
```
