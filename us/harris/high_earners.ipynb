```python
def run_simulation(year, baseline, reformed=None):
    if reformed is None:
        return 0
    baseline_income = baseline.calculate("household_net_income", period=year)
    reformed_income = reformed.calculate("household_net_income", period=year)
    return (reformed_income - baseline_income).sum()
```

```python
def create_reform_comparison():
    import os
    
    # Initialize DataFrame
    years = range(2025, 2035)
    columns = ['NIIT', 'Medicare', 'Top Income Rate', '28% Capital Gains', 'Total']
    df = pd.DataFrame(0, index=list(years), columns=columns)
    
    # Check for existing CSV
    csv_path = 'reform_comparison_results.csv'
    if os.path.exists(csv_path):
        existing_df = pd.read_csv(csv_path, index_col=0)
        df.update(existing_df)
        print("Loaded existing results from CSV")
    
    # Create simulations
    baseline = Microsimulation(dataset='enhanced_cps_2024')
    reforms = {
        'NIIT': Microsimulation(reform=niit_reform, dataset='enhanced_cps_2024'),
        'Medicare': Microsimulation(reform=medicare_reform, dataset='enhanced_cps_2024'),
        'Top Income Rate': Microsimulation(reform=income_tax_reform, dataset='enhanced_cps_2024'),
        '28% Capital Gains': Microsimulation(reform=capital_gains_reform, dataset='enhanced_cps_2024'),
        'Total': Microsimulation(reform=combined_reform, dataset='enhanced_cps_2024')
    }
    
    # Process each year and reform
    for year in years:
        for reform_name, reform_sim in reforms.items():
            # Skip if value already exists
            if pd.notna(df.at[year, reform_name]):
                print(f"Skipping {year} {reform_name} - already calculated")
                continue
            
            print(f"Processing {year} {reform_name}")
            try:
                baseline_income = baseline.calculate("household_net_income", period=year)
                reformed_income = reform_sim.calculate("household_net_income", period=year)
                df.at[year, reform_name] = (reformed_income - baseline_income).sum() / 1e9
                
                # Save after each calculation
                df.to_csv(csv_path)
                print(f"Saved progress to {csv_path}")
            except Exception as e:
                print(f"Error processing {year} {reform_name}: {str(e)}")
    
    # Calculate final totals
    df.loc['2025-34'] = df.loc[years].sum()
    df = df.round(1)
    df.index = [str(year) for year in df.index]
    
    return df
```
