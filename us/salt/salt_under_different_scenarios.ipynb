```python
def generate_tax_reform_csv(
    baseline, tcja_reform, tcja_25k_reform, output_file="tax_reform_comparison.csv"
):
    """
    Generate a CSV file with tax data for 1000 sampled households across three reform scenarios
    Uses fully vectorized operations with original variable names as columns
    """
    import pandas as pd
    import numpy as np

    # Define the reforms
    reforms = {
        "Current law": baseline,
        "Current policy (TCJA extended)": tcja_reform,
        "Current policy (TCJA extended) with $25k SALT cap": tcja_25k_reform,
    }

    # Get all household IDs and sample 1000 households
    household_ids = baseline.calculate("household_id", map_to="household", period=2026)
    num_households = len(household_ids)
    sample_indices = np.random.choice(num_households, size=1000, replace=False)

    # Variables to extract
    variables = [
        "is_married",
        "tax_unit_children",
        "state_code",
        "adjusted_gross_income",
        "real_estate_taxes",
        "state_and_local_sales_or_income_tax",
        "salt",
        "income_tax",
    ]

    # Create a dataframe for each reform and concatenate them
    dfs = []

    for reform_name, reform_sim in reforms.items():
        # Get all variables at once, only for sampled households
        data = {
            var: reform_sim.calculate(var, map_to="household", period=2026).round(0)[
                sample_indices
            ]
            for var in variables
        }

        # Add index and reform columns
        data["index"] = sample_indices
        data["reform"] = reform_name

        # Create dataframe for this reform
        reform_df = pd.DataFrame(data)
        dfs.append(reform_df)

    # Concatenate all dataframes and sort to keep households together
    df = pd.concat(dfs, ignore_index=True)
    df = df.sort_values(["index", "reform"])

    # Write to CSV
    df.to_csv(output_file, index=False)

    return df
```

```python
df = generate_tax_reform_csv(baseline, tcja_reform, tcja_25k_reform)
```

```python
# First get the static household data from one reform only
household_data = df[df["reform"] == "Current law"].copy()

# Keep only one set of household characteristics
static_cols = [
    "index",
    "is_married",
    "tax_unit_children",
    "state_code",
    "adjusted_gross_income",
    "real_estate_taxes",
    "state_and_local_sales_or_income_tax",
    "salt",
]

# Create base dataframe with one row per household
wide_df = household_data[static_cols].copy()

# Pivot the income tax values
tax_pivot = df.pivot(index="index", columns="reform", values="income_tax")

# Rename columns to avoid spaces
tax_pivot.columns = [
    f"income_tax_{col.lower().replace(' ', '_').replace('(', '').replace(')', '').replace('$', '')}"
    for col in tax_pivot.columns
]

# Reset index to join with wide_df
tax_pivot.reset_index(inplace=True)

# Merge with household data
wide_df = wide_df.merge(tax_pivot, on="index")

# Calculate tax differences
wide_df["tax_diff_tcja_vs_current"] = (
    wide_df["income_tax_current_policy_tcja_extended"]
    - wide_df["income_tax_current_law"]
)

wide_df["tax_diff_25k_vs_tcja"] = (
    wide_df["income_tax_current_policy_tcja_extended_with_25k_salt_cap"]
    - wide_df["income_tax_current_policy_tcja_extended"]
)

# Export to CSV
wide_df.to_csv("tax_reform_wide.csv", index=False)
```
