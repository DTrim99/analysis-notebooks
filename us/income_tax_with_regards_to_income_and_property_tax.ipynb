```python


# Set up the different state-income-real estate tax mappings
REAL_ESTATE_TAX_MAPPING = {
    "CA": {
        250_000: 9_940,
        500_000: 19_880,
        750_000: 29_820,
        1_000_000: 39_760
    },
    "PA": {
        250_000: 18_900,
        500_000: 37_800,
        750_000: 56_700,
        1_000_000: 75_600
    },
    "NJ": {
        250_000: 31_220,
        500_000: 62_440,
        750_000: 93_660,
        1_000_000: 124_880
    },
    "NY": {
        250_000: 22_400,
        500_000: 44_800,
        750_000: 67_200,
        1_000_000: 89_600
    }
}

# Create a function to run simulations for a specific state, income, and reform
def run_simulation(state, income, real_estate_tax, reform=None):
    situation = {
        "people": {
            "you": {
                "age": {
                    "2026": 40
                },
                "employment_income": {
                    "2026": income
                },
                "real_estate_taxes": {
                    "2026": real_estate_tax
                }
            },
            "your partner": {
                "age": {
                    "2026": 40
                },
                "employment_income": {
                    "2026": 0
                }
            }
        },
        "families": {
            "your family": {
                "members": ["you", "your partner"]
            }
        },
        "marital_units": {
            "your marital unit": {
                "members": ["you", "your partner"]
            }
        },
        "tax_units": {
            "your tax unit": {
                "members": ["you", "your partner"]
            }
        },
        "spm_units": {
            "your household": {
                "members": ["you", "your partner"]
            }
        },
        "households": {
            "your household": {
                "members": ["you", "your partner"],
                "state_name": {
                    "2026": state
                }
            }
        }
    }
    
    sim = Simulation(situation=situation, reform=reform)
    # Calculate and return income tax
    income_tax = sim.calculate("income_tax", "2026")
    return income_tax
```

```python

# Create results DataFrame
results = []

# Taxable income levels
WAGES = [250_000, 500_000, 750_000, 1_000_000]

# Run simulations for each state, income level, and reform
for state in REAL_ESTATE_TAX_MAPPING:
    for income in WAGES:
        real_estate_tax = REAL_ESTATE_TAX_MAPPING[state][income]
        
        # Run baseline simulation
        baseline_tax = run_simulation(state, income, real_estate_tax)
        
        # Run TCJA simulation
        tcja_tax = run_simulation(state, income, real_estate_tax, tcja_reform)
        
        # Run TCJA 50k simulation
        tcja_50k_tax = run_simulation(state, income, real_estate_tax, tcja_50k_reform)

        # Run TCJA 30k simulation
        tcja_30k_tax = run_simulation(state, income, real_estate_tax, tcja_30k_reform)
        
        # Add to results
        results.append({
            "State": state,
            "Wages": income,
            "Property Tax": real_estate_tax,
            "Baseline Tax": baseline_tax,
            "TCJA Tax": tcja_tax,
            "TCJA 50k Tax": tcja_50k_tax,
            "TCJA 30k Tax": tcja_30k_tax
        })

# Convert to DataFrame for easier analysis
results_df = pd.DataFrame(results)
```
